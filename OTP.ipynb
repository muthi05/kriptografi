```python
cipher = input("Masukkan ciphertext (huruf besar, tanpa spasi): ").upper()
plaintext = input("Masukkan plaintext (huruf besar, tanpa spasi): ").upper()

import string

alfabet = string.ascii_uppercase
to_num = {c: i for i, c in enumerate(alfabet)}
to_char = {i: c for i, c in enumerate(alfabet)}

def text_to_num(text):
    return [to_num[c] for c in text]

def num_to_text(nums):
    return ''.join(to_char[n] for n in nums)

def find_key(ciphertext, plaintext):
    n = min(len(ciphertext), len(plaintext))
    ciphertext = ciphertext[:n]
    plaintext = plaintext[:n]
    C = text_to_num(ciphertext)
    P = text_to_num(plaintext)
    K = [(c - p) % 26 for c, p in zip(C, P)]
    return num_to_text(K)

key = find_key(cipher, plaintext)
print("Key:", key)
```
